```python
data={"Product_Name":[],"Price":[],"Rating":[],"Reviews":[],"Website":[],"Image_link":[]}
def scrape(urlM):
   
    if urlM==urlA:
        driver=webdriver.Chrome(executable_path="C:\webdrivers\chromedriver.exe")
        driver.get(urlA)
        page_soupA=soup(driver.page_source,"html.parser")
        #print(page_soupA)
        containerA=page_soupA.find_all('div',{'data-component-type':"s-search-result"})
        conA=containerA[0]
        soup.prettify(conA)
        print("-------------------------------------------------")

        for co in containerA[0:6]:
            atag=co.h2.a
            print(atag.text)
            data["Product_Name"].append(atag.text)
            data["Website"].append("AMAZON.IN")
        print("-------------------------------------------------")

        for pr in containerA[0:6]:
                prA=pr.find('span','a-price')

                try:
                    priceA=prA('span','a-offscreen')
                    print(priceA[0].text)
                    data["Price"].append(priceA[0].text)

                except(TypeError):
                    print()
                    data["Price"].append("NO PRICE LISTED")
        print("-------------------------------------------------")

        for rat in containerA[0:6]:
            try:
                rating=rat.i.text
                print(rating)
                data["Rating"].append(rating)
            except(AttributeError):
                print()
                data["Rating"].append("NO RATINGS")

        print("-------------------------------------------------")
        for rew in containerA[0:6]:
            try:
                rewC=rew.find('span',{'class':"a-size-base s-underline-text"})
                print(rewC.text)
                data["Reviews"].append(rewC.text)
            except(AttributeError):
                print()
                data["Reviews"].append("NO REVIEWS")
        
        for link in containerA[0:6]:
            try:
                img=link.find('img',{"class":"s-image"}).attrs["src"]
                print(img)
                data["Image_link"].append(img)
            except(Exception):
                print("No Image avl")
                
                
                
                
                
    elif urlM==url:
        try:
            uClient=urlopen(url)
            page_html=uClient.read()
            uClient.close()
            page_soup=soup(page_html,"html.parser")
            container=page_soup.find_all("div",{"class":"_2kHMtA"})
            con=container[0]
            soup.prettify(con)
            print("-------------------------------------------------")

            for co in container[0:6]:
                product_name=co.findAll("div",{"class":"_4rR01T"})
                print(product_name[0].text)
                data["Product_Name"].append(product_name[0].text)
                data["Website"].append("FLIPKART.COM")
            print("-------------------------------------------------")

            for co in container[0:6]:
                try:
                    s_price=co.findAll("div",{"class":"_30jeq3 _1_WHN1"})
                    print(s_price[0].text)
                    data["Price"].append(s_price[0].text)
                    print("-------------------------------------------------")
                except(Exception):
                    data["Price"].append("No price available")

                    
                    

            for co in container[0:6]:

                try:
                    rating=co.findAll("span",{"class":"_2_R_DZ"})
                    print(rating[0].text)
                    data["Reviews"].append(rating[0].text)
                except(IndexError):
                    print()
                    data["Reviews"].append("NO REVIEWS")
            print("-------------------------------------------------")

            for co in container[0:6]:

                try:
                    star=co.findAll("div",{"class":"_3LWZlK"})
                    print(star[0].text)
                    data["Rating"].append(star[0].text)
                except(IndexError):
                    print()
                    data["Rating"].append("NO RATINGS")
            
            for co in container[0:6]:
                try:
                    img=co.find_all("img",{"class":"_396cs4 _3exPp9"},"src")
                    print(img)
                except(Exception):
                    print()
            for co in container[0:6]:
                try:
                    img=co.find("img",{"class":"_396cs4 _3exPp9"}).attrs["src"]
                    print(img)
                    data["Image_link"].append(img)
                except(Exception):
                    data["Image_link"].append("No Image")


            
        except:
            print("NO data available")
            

keyword=input("Enter product to search: ")
keyword=keyword.replace(" ","+")

url="https://www.flipkart.com/search?q="+keyword
urlA="https://www.amazon.in/s?k="+keyword


print("Connecting to amazon")
scrape(urlA)
print("Connecting to Flipkart")
scrape(url)
            

```
